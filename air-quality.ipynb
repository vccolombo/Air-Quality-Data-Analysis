```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("AirQualityUCI.csv", delimiter=';', decimal=",")
df
```

```python
df.describe()
```

```python
df.info()
```

```python
# remover colunas adicionais que nÃ£o tem nenhuma entrada
df = df.drop(columns=['Unnamed: 15', 'Unnamed: 16'])

# remover data e horÃ¡rio
df = df.drop(columns=["Date", "Time"])
```

```python
# descobrir quantos valores sÃ£o nulos
df.isnull().sum()
```

```python
df = df.dropna() # retirar todas as linhas que tenham entradas nulas
# Verificar se retirou os nulos
df.isnull().sum()
```

```python
# Visualizar quais colunas tem dados faltando (-200 de acordo com a documentaÃ§Ã£o)
df = df.replace(-200, np.nan)

df.isnull().sum()
```

```python
# Retirar linhas em que AH Ã© nan
df = df.dropna(subset=["AH"])
df.isnull().sum()
```

```python
# Remover coluna com quase todos os dados faltando
df = df.drop(columns=["NMHC(GT)"])

# substituir o resto pela mÃ©dia
col = "CO(GT)"
df[col] = df[col].fillna((df[col].mean()))

# col = "PT08.S1(CO)"
# df[col] = df[col].fillna((df[col].mean()))

# col = "C6H6(GT)"
# df[col] = df[col].fillna((df[col].mean()))

# col = "PT08.S2(NMHC)"
# df[col] = df[col].fillna((df[col].mean()))

col = "NOx(GT)"
df[col] = df[col].fillna((df[col].mean()))

# col = "PT08.S3(NOx)"
# df[col] = df[col].fillna((df[col].mean()))

col = "NO2(GT)"
df[col] = df[col].fillna((df[col].mean()))

# col = "PT08.S4(NO2)"
# df[col] = df[col].fillna((df[col].mean()))

# col = "PT08.S5(O3)"
# df[col] = df[col].fillna((df[col].mean()))

# col = "T"
# df[col] = df[col].fillna((df[col].mean()))

df.isnull().sum()
```

```python
df.describe()
```

```python
# Remove possible outliers
q = df.quantile(0.99)
df = df[df < q]

q = df.quantile(0.01)
df = df[df > q]

df = df.dropna()

df
```

```python
normalized_df=(df-df.mean())/df.std()
normalized_df["AH"] = df["AH"]

normalized_df
```

```python
import seaborn as sns
# calculate the correlation matrix
corr = normalized_df.corr()
corr
```

```python
# plot the heatmap
sns.heatmap(corr, 
        xticklabels=corr.columns,
        yticklabels=corr.columns)
```

```python
# Temperature
import matplotlib.pyplot as plt
x = normalized_df["T"]
y = df["AH"]
plt.scatter(x, y)
plt.xlabel("Temperature (celsius)")
plt.ylabel("Absolute Humidity")
```
