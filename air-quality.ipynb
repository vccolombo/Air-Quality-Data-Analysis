```python
# remover colunas adicionais que não tem nenhuma entrada
df = df.drop(columns=['Unnamed: 15', 'Unnamed: 16'])
```

```python
# descobrir quantos valores são nulos
df.isnull().sum()
```

```python
df = df.dropna() # retirar todas as linhas que tenham entradas nulas
# Verificar se retirou os nulos
df.isnull().sum()
```

```python
# Visualizar quais colunas tem dados faltando (-200 de acordo com a documentação)
df = df.replace(-200, np.nan)

df.isnull().sum()
```

```python
# Retirar linhas em que AH é nan
df = df.dropna(subset=["AH"])
df.isnull().sum()
```
