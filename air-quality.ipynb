```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("AirQualityUCI.csv", delimiter=';', decimal=",")
df
```

```python
# remover colunas adicionais que nÃ£o tem nenhuma entrada
df = df.drop(columns=['Unnamed: 15', 'Unnamed: 16'])

# remover data e horÃ¡rio
df = df.drop(columns=["Date", "Time"])
```

```python
# descobrir quantos valores sÃ£o nulos
df.isnull().sum()
```

```python
df = df.dropna() # retirar todas as linhas que tenham entradas nulas
# Verificar se retirou os nulos
df.isnull().sum()
```

```python
# Visualizar quais colunas tem dados faltando (-200 de acordo com a documentaÃ§Ã£o)
df = df.replace(-200, np.nan)

df.isnull().sum()
```

```python
# Retirar linhas em que AH Ã© nan
df = df.dropna(subset=["AH"])
df.isnull().sum()
```

```python
# Remover coluna com quase todos os dados faltando
df = df.drop(columns=["NMHC(GT)"])

# substituir o resto pela mÃ©dia
col = "CO(GT)"
df[col] = df[col].fillna((df[col].mean()))

# col = "PT08.S1(CO)"
# df[col] = df[col].fillna((df[col].mean()))

# col = "C6H6(GT)"
# df[col] = df[col].fillna((df[col].mean()))

# col = "PT08.S2(NMHC)"
# df[col] = df[col].fillna((df[col].mean()))

col = "NOx(GT)"
df[col] = df[col].fillna((df[col].mean()))

# col = "PT08.S3(NOx)"
# df[col] = df[col].fillna((df[col].mean()))

col = "NO2(GT)"
df[col] = df[col].fillna((df[col].mean()))

# col = "PT08.S4(NO2)"
# df[col] = df[col].fillna((df[col].mean()))

# col = "PT08.S5(O3)"
# df[col] = df[col].fillna((df[col].mean()))

# col = "T"
# df[col] = df[col].fillna((df[col].mean()))

df.isnull().sum()
```

```python
df.describe()
```

```python
# Remove possible outliers
q = df.quantile(0.99)
df = df[df < q]

q = df.quantile(0.01)
df = df[df > q]

df = df.dropna()

df
```

```python
normalized_df=(df-df.mean())/df.std()
normalized_df["RH"] = df["RH"]

normalized_df
```

```python
import seaborn as sns
# calculate the correlation matrix
corr = normalized_df.corr()
corr
```

```python
# plot the heatmap
import matplotlib.pyplot as plt
sns.heatmap(corr,
        xticklabels=corr.columns,
        yticklabels=corr.columns)
```

```python
import matplotlib.pyplot as plt
g = sns.PairGrid(normalized_df, y_vars="AH", x_vars=[c for c in df.columns if c != "RH"])
g.map(plt.scatter);
```

```python
# Escolher quais features vamos manter
target = normalized_df["RH"]

df_features = normalized_df.drop(columns=["AH", "RH"])
final_features = df_features[["T", "NOx(GT)", "PT08.S5(O3)", "NO2(GT)", "PT08.S2(NMHC)", "C6H6(GT)"]]


final_features
```

```python
X = final_features
Y = target
```

```python
from sklearn.linear_model import LinearRegression

lm = LinearRegression()
lm.fit(X, Y)
Yhat = lm.predict(X)
```

```python
from sklearn.metrics import mean_squared_error
import numpy as np
print("R2 = " + str(lm.score(X, Y)))
print("MSE = " + str(mean_squared_error(Y,Yhat)))
```

```python
from sklearn.preprocessing import PolynomialFeatures

pr = PolynomialFeatures(degree=3, include_bias=False)
x_poly = pr.fit_transform(X)

model = LinearRegression()
model.fit(x_poly, Y)
y_poly_pred = model.predict(x_poly)
```

```python
from sklearn.metrics import mean_squared_error
import numpy as np

print("R2 = " + str(model.score(x_poly, Y)))
print("MSE = " + str(mean_squared_error(Y,y_poly_pred)))
```
